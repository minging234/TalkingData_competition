## App distribution

```python
app = X['app'].value_counts()
```

```python
app_top10 = app[:10]
print("Top 10 apps comprise {} of total data".format((app.cumsum()/app.sum()).iloc[10] * 100))
```

```python
bins = np.arange(app_top10.size) + .5
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(bins, app_top10)
# ax.set_yscale('log', nonposy='clip')
ax.tick_params(labelsize=13)
ax.set_xlabel('App id for marketing', fontsize=13);
ax.set_xticks(bins)
ax.set_xticklabels(app_top10.index.values)
ax.set_ylabel('Frequency in Millions', fontsize=13);
ax.set_title('Histogram of Top 10 App ID for first 100 Million Rows', fontsize=15);
```

## Click Time EDA

```python
days = X['click_time'].dt.dayofweek

days = days.value_counts()

days
```

```python
bins = np.arange(4) + .5
labels = ['Wednesday', 'Tuesday', 'Thursday', 'Monday']

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(bins, days)
# ax.set_yscale('log', nonposy='clip')
ax.tick_params(labelsize=13)
ax.set_xticks(bins)
ax.set_xticklabels(labels)
ax.set_xlabel('Day of Week', fontsize=13);
ax.set_ylabel('Frequency', fontsize=13);
ax.set_title('Histogram of Days of clicks', fontsize=15);
```

```python
hour = X['click_time'].dt.hour
```

```python
hour = hour.value_counts()
```

```python
hour = hour.sort_index()
```

```python
bins = np.arange(len(hour))

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(hour.index, hour, marker='o')
# ax.set_yscale('log', nonposy='clip')
ax.tick_params(labelsize=13)
ax.set_xticks(bins)
ax.set_xticklabels(hour.index)
ax.set_xlabel('Hour of Day', fontsize=13);
ax.set_ylabel('Frequency', fontsize=13);
ax.set_title('Histogram of Days of clicks', fontsize=15);
```

```python
days = X['click_time'].dt.dayofweek
```

```python
pd.unique(days)
```

```python
hour = X['click_time'].dt.hour
```

```python
fig, ax = plt.subplots(4, 1, figsize=(10, 16))

for i in np.arange(4, dtype=np.int):
    mask = days[days == i]
    y_hour = hour[mask]
    y_hour = y_hour.value_counts()
    y_hour = y_hour.sort_index()
    ax[i].plot(y_hour.index, y_hour)
```
